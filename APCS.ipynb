```python
#@title h082. 2. 贏家預測
n,m=map(int,input().split())
s=[0]+[int(x) for x in input().split()]
t=[0]+[int(x) for x in input().split()]
idx=[int(x) for x in input().split()]
fail=[0]*(n+1)
while len(idx)>1:
    win=[]
    lose=[]
    for i in range(0,len(idx)-1,2):
        a,b,c,d=s[idx[i]],t[idx[i]],s[idx[i+1]],t[idx[i+1]]
        if a*b>=c*d:
            s[idx[i]]=a+c*d//(2*b)
            t[idx[i]]=b+c*d//(2*a)
            s[idx[i+1]]=c+c//2
            t[idx[i+1]]=d+d//2
            fail[idx[i+1]]+=1
            if fail[idx[i]]<m:
                win.append(idx[i])
            if fail[idx[i+1]]<m:
                lose.append(idx[i+1])
        else:
            s[idx[i+1]]=c+a*b//(2*d)
            t[idx[i+1]]=d+a*b//(2*c)
            s[idx[i]]=a+a//2
            t[idx[i]]=b+b//2
            fail[idx[i]]+=1
            if fail[idx[i+1]]<m:
                win.append(idx[i+1])
            if fail[idx[i]]<m:
                lose.append(idx[i])
    if len(idx)%2==1:
        win.append(idx[-1])
    idx=win+lose
print(idx[0])
```

```python
#@title g276. 2. 魔王迷宮
n,m,k=map(int,input().split())
l=[[int(i) for i in input().split()] for _ in range(k)]
board=set()
d2=[]
alive=k
while alive>0:
    d1=set()
    for i in range(len(l)):
        if i in d2:
            continue
        board.add(tuple([l[i][0],l[i][1]]))
    for i in range(len(l)):
        if i in d2:
            continue
        l[i][0]+=l[i][2]
        l[i][1]+=l[i][3]
        if not(0<=l[i][0]<n) or not(0<=l[i][1]<m):
            d2.append(i)
            alive-=1
        elif tuple([l[i][0],l[i][1]]) in board:
            d1.add(tuple([l[i][0],l[i][1]]))
            d2.append(i)
            alive-=1
    for j in d1:
        board.remove(j)
print(len(board))
```

```python
from collections import Counter

# 創建範例數據
data = ['apple', 'banana', 'apple', 'cherry', 'apple', 'banana', 'date']
counter = Counter(data)

print("原始數據:", data)
print("Counter 對象:", counter)
print("=" * 50)

# 1. most_common() 不帶參數 - 返回所有項目，按出現次數降序排列
print("1. most_common() 不帶參數:")
all_common = counter.most_common()
print("返回結果:", all_common)
print("數據類型:", type(all_common))
print("每個元素類型:", type(all_common[0]))
print()

# 2. most_common(1) - 只返回出現次數最多的 1 個項目
print("2. most_common(1) - 返回最多的 1 個:")
top_1 = counter.most_common(1)
print("返回結果:", top_1)
print("數據類型:", type(top_1))
print("這是一個列表，包含 1 個元組")
print()

# 3. most_common(2) - 返回出現次數最多的前 2 個項目
print("3. most_common(2) - 返回最多的前 2 個:")
top_2 = counter.most_common(2)
print("返回結果:", top_2)
print()

# 4. most_common(3) - 返回出現次數最多的前 3 個項目
print("4. most_common(3) - 返回最多的前 3 個:")
top_3 = counter.most_common(3)
print("返回結果:", top_3)
print()

print("=" * 50)
print("解析 most_common(1) 的返回值:")
print("=" * 50)

most_common_1 = counter.most_common(1)
print("most_common(1) 返回:", most_common_1)
print("返回值是列表:", isinstance(most_common_1, list))
print("列表長度:", len(most_common_1))
print()

# 提取第一個（也是唯一一個）元組
first_tuple = most_common_1[0]
print("第一個元組:", first_tuple)
print("元組結構: (項目, 出現次數)")
print()

# 提取項目和次數
item = first_tuple[0]
count = first_tuple[1]
print("項目:", item)
print("出現次數:", count)
print()

# 這就是為什麼在原始函數中使用 counter.most_common(1)[0][0]
print("完整提取過程:")
print("counter.most_common(1)      ->", counter.most_common(1))
print("counter.most_common(1)[0]   ->", counter.most_common(1)[0])
print("counter.most_common(1)[0][0] ->", counter.most_common(1)[0][0])
print()

print("=" * 50)
print("不同情況的範例:")
print("=" * 50)

# 範例 1: 有明確最多次的項目
example1 = [1, 2, 2, 2, 3, 3]
counter1 = Counter(example1)
print("範例 1:", example1)
print("Counter:", counter1)
print("most_common():", counter1.most_common())
print("most_common(1):", counter1.most_common(1))
print("最多次項目:", counter1.most_common(1)[0][0])
print()

# 範例 2: 多個項目並列最多次
example2 = [1, 1, 2, 2, 3]
counter2 = Counter(example2)
print("範例 2 (並列情況):", example2)
print("Counter:", counter2)
print("most_common():", counter2.most_common())
print("most_common(1):", counter2.most_common(1))
print("注意: most_common(1) 只返回其中一個並列的項目")
print("最多次項目:", counter2.most_common(1)[0][0])
print()

# 範例 3: 獲取並列的所有最多次項目
print("如果要獲取所有並列的最多次項目:")
max_count = counter2.most_common(1)[0][1]  # 獲取最高出現次數
all_max_items = [item for item, count in counter2.items() if count == max_count]
print("最高出現次數:", max_count)
print("所有並列最多次的項目:", all_max_items)
print()

print("=" * 50)
print("實用技巧:")
print("=" * 50)

# 技巧 1: 解構賦值
item, count = counter.most_common(1)[0]
print("技巧 1 - 解構賦值:")
print(f"item, count = counter.most_common(1)[0]")
print(f"結果: item = '{item}', count = {count}")
print()

# 技巧 2: 獲取前 N 個最多次的項目
print("技巧 2 - 獲取前 N 個:")
for i, (item, count) in enumerate(counter.most_common(3), 1):
    print(f"第 {i} 名: '{item}' 出現 {count} 次")
print()

# 技巧 3: 檢查是否有並列情況
print("技巧 3 - 檢查並列:")
top_2_items = counter.most_common(2)
if len(top_2_items) >= 2 and top_2_items[0][1] == top_2_items[1][1]:
    print("有並列第一名！")
else:
    print("有明確的第一名")
print()

print("=" * 50)
print("常見錯誤和注意事項:")
print("=" * 50)

print("❌ 錯誤: 忘記 most_common() 返回的是列表")
print("# counter.most_common(1) 不能直接當作元組使用")
print()

print("❌ 錯誤: 空列表情況")
empty_counter = Counter([])
print("空 Counter:", empty_counter)
print("empty_counter.most_common(1):", empty_counter.most_common(1))
print("需要先檢查列表是否為空！")
print()

print("✅ 正確的安全用法:")
def safe_get_most_common(lst):
    if not lst:
        return None
    counter = Counter(lst)
    return counter.most_common(1)[0][0]

print("safe_get_most_common([]):", safe_get_most_common([]))
print("safe_get_most_common(['a', 'b', 'a']):", safe_get_most_common(['a', 'b', 'a']))
```

```python
from collections import Counter

print("=" * 60)
print("什麼是 Counter？")
print("=" * 60)

print("Counter 是 Python collections 模組中的一個特殊字典子類")
print("專門用於計數可雜湊（hashable）對象的出現次數")
print()

# 1. 基本創建方式
print("1. Counter 的不同創建方式:")
print("-" * 30)

# 從列表創建
data_list = ['a', 'b', 'c', 'a', 'b', 'a']
counter_from_list = Counter(data_list)
print(f"從列表創建: Counter({data_list})")
print(f"結果: {counter_from_list}")
print()

# 從字串創建
data_string = "hello"
counter_from_string = Counter(data_string)
print(f"從字串創建: Counter('{data_string}')")
print(f"結果: {counter_from_string}")
print()

# 從字典創建
data_dict = {'a': 3, 'b': 2, 'c': 1}
counter_from_dict = Counter(data_dict)
print(f"從字典創建: Counter({data_dict})")
print(f"結果: {counter_from_dict}")
print()

# 從關鍵字參數創建
counter_from_kwargs = Counter(a=3, b=2, c=1)
print(f"從關鍵字參數創建: Counter(a=3, b=2, c=1)")
print(f"結果: {counter_from_kwargs}")
print()

# 空的 Counter
empty_counter = Counter()
print(f"空 Counter: Counter()")
print(f"結果: {empty_counter}")
print()

print("=" * 60)
print("Counter 內部是如何工作的？")
print("=" * 60)

data = ['apple', 'banana', 'apple', 'cherry', 'apple', 'banana']
print(f"原始數據: {data}")
print()

print("步驟解析:")
print("1. Counter 遍歷每個元素")
print("2. 為每個元素建立計數")
print("3. 相同元素的計數累加")
print()

# 手動模擬 Counter 的工作過程
manual_count = {}
print("手動計數過程:")
for i, item in enumerate(data):
    if item in manual_count:
        manual_count[item] += 1
    else:
        manual_count[item] = 1
    print(f"  步驟 {i+1}: 處理 '{item}' -> {dict(manual_count)}")

print()
print("最終手動計數結果:", manual_count)

# 用 Counter 做同樣的事
counter_result = Counter(data)
print("Counter 自動計數結果:", dict(counter_result))
print("兩者結果相同!", manual_count == dict(counter_result))
print()

print("=" * 60)
print("Counter 與普通字典的比較")
print("=" * 60)

# 普通字典 vs Counter
normal_dict = {}
counter_dict = Counter()

print("處理不存在的鍵時的差異:")
print()

print("普通字典:")
try:
    value = normal_dict['不存在的鍵']
except KeyError as e:
    print(f"  訪問不存在的鍵會拋出 KeyError: {e}")

print()
print("Counter:")
value = counter_dict['不存在的鍵']
print(f"  訪問不存在的鍵返回: {value}")
print("  Counter 對不存在的鍵返回 0，不會拋出錯誤")
print()

print("=" * 60)
print("Counter 的特殊屬性和方法")
print("=" * 60)

counter = Counter(['a', 'b', 'c', 'a', 'b', 'a'])
print(f"範例 Counter: {counter}")
print()

print("1. 基本屬性:")
print(f"   類型: {type(counter)}")
print(f"   是否為字典的子類: {isinstance(counter, dict)}")
print(f"   長度: {len(counter)}")
print(f"   鍵: {list(counter.keys())}")
print(f"   值: {list(counter.values())}")
print(f"   項目: {list(counter.items())}")
print()

print("2. 特殊方法:")

# most_common()
print(f"   most_common(): {counter.most_common()}")
print(f"   most_common(2): {counter.most_common(2)}")

# elements()
print(f"   elements(): {list(counter.elements())}")

# total()
print(f"   total(): {counter.total()}")  # Python 3.10+
print()

print("3. 數學運算:")
counter1 = Counter(['a', 'b', 'c', 'a'])
counter2 = Counter(['a', 'b', 'b', 'd'])

print(f"   counter1: {counter1}")
print(f"   counter2: {counter2}")
print(f"   相加: {counter1 + counter2}")
print(f"   相減: {counter1 - counter2}")
print(f"   交集(&): {counter1 & counter2}")
print(f"   聯集(|): {counter1 | counter2}")
print()

print("=" * 60)
print("實際使用範例")
print("=" * 60)

# 範例 1: 統計文字中字母出現次數
text = "hello world"
letter_count = Counter(text)
print(f"文字: '{text}'")
print(f"字母統計: {letter_count}")
print(f"最常見的字母: {letter_count.most_common(1)[0][0]}")
print()

# 範例 2: 統計單詞出現次數
sentence = "the quick brown fox jumps over the lazy dog the fox"
words = sentence.split()
word_count = Counter(words)
print(f"句子: '{sentence}'")
print(f"單詞統計: {word_count}")
print(f"最常見的單詞: {word_count.most_common(1)[0]}")
print()

# 範例 3: 統計數字出現次數
numbers = [1, 2, 3, 1, 2, 1, 4, 5, 1]
number_count = Counter(numbers)
print(f"數字列表: {numbers}")
print(f"數字統計: {number_count}")
print(f"出現最多的數字: {number_count.most_common(1)[0][0]}")
print()

print("=" * 60)
print("Counter 的進階用法")
print("=" * 60)

# 更新計數
counter = Counter(['a', 'b', 'c'])
print(f"初始 counter: {counter}")

# update() 方法
counter.update(['a', 'b', 'a'])
print(f"update(['a', 'b', 'a']) 後: {counter}")

# subtract() 方法
counter.subtract(['a', 'b'])
print(f"subtract(['a', 'b']) 後: {counter}")
print()

# 計數可以是負數
counter_with_negative = Counter()
counter_with_negative.update(['a', 'b'])
counter_with_negative.subtract(['a', 'a', 'b', 'b', 'c'])
print(f"負數計數範例: {counter_with_negative}")
print()

print("=" * 60)
print("常見錯誤和注意事項")
print("=" * 60)

print("❌ 錯誤 1: 嘗試計數不可雜湊的對象")
try:
    # 列表是不可雜湊的，不能作為字典的鍵
    bad_counter = Counter([['a', 'b'], ['c', 'd']])
except TypeError as e:
    print(f"   TypeError: {e}")

print()
print("✅ 解決方案: 轉換為可雜湊類型")
# 將列表轉換為元組
good_counter = Counter([tuple(['a', 'b']), tuple(['c', 'd']), tuple(['a', 'b'])])
print(f"   正確用法: {good_counter}")
print()

print("❌ 錯誤 2: 混淆 Counter 和普通字典")
print("   Counter['不存在的鍵'] 返回 0，不會報錯")
print("   但在邏輯判斷時要注意這個特性")
print()

print("✅ 最佳實踐:")
print("1. 用 Counter 統計計數")
print("2. 用 most_common() 獲取排序結果")
print("3. 記住 Counter 是字典的子類，可以使用字典的所有方法")
print("4. 對於大數據集，Counter 比手動計數更高效")
print()

print("=" * 60)
print("效能比較")
print("=" * 60)

import time
import random

# 創建大型測試數據
large_data = [random.randint(1, 100) for _ in range(100000)]

# 方法 1: 使用 Counter
start_time = time.time()
counter_result = Counter(large_data)
counter_time = time.time() - start_time

# 方法 2: 手動字典計數
start_time = time.time()
manual_dict = {}
for item in large_data:
    manual_dict[item] = manual_dict.get(item, 0) + 1
manual_time = time.time() - start_time

print(f"Counter 方法: {counter_time:.4f} 秒")
print(f"手動計數方法: {manual_time:.4f} 秒")
print(f"Counter 比手動方法快: {manual_time/counter_time:.2f} 倍")
```
